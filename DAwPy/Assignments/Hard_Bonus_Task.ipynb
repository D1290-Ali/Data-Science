
**HARD BONUS TASK: Create a table for counts for each day of the week that people arrived. (E.g. 5000 arrivals were on a Monday, 3000 were on a Tuesday, etc..)**

Veri setinde otele giriş tarihini gösteren bir sütun yok. Bunun yerine gün, ay ve yıl ayrı ayrı verilmiş. Öncelikle bu üç sutündan bir tarih sütunu oluşturmamız gerekiyor. Aşağıdaki kod ile veri setini yüklerken bu üç sütunu birleştiriyoruz. "parse_date="parametresi liste içinde yer alan üç sütunu "arrival_date"sütunu olarak tarih formatında birleştiriyor. Diğer üç sütun artık data setinde yer almıyor.

```python
hotels = pd.read_csv( '/content/drive/MyDrive/hotel_booking_data.csv', parse_dates={ 'arrival_date': ['arrival_date_year','arrival_date_month','arrival_date_day_of_month']})
```

Artık veri setinde otele giriş tarihlerini gösteren "arrival_date" sütunu var. Ancak bu tarihlerin haftanın hangi gününe denk geldiğini gösteren bir veri yok. Bu sütunu elde etmek için aşağıda yer alan "dt.day_name()" metodunu kullanıyoruz. Artık haftanın günleri "day_name" sütunu olarak veri tabanında yer alıyor.

```python
hotels.columns  # çalıştırdığınızda aşağıdaki ilk sütun
```

```python
hotels["day_name"]=hotels["arrival_date"].dt.day_name()
hotels["day_name"]
```

```python
hotels.columns   # çalıştırdığınızda aşağıdaki son sütun
```

Artık "day_name" sütununu kullanarak "value_counts" yaptığımızda herbir güne denk gelen giriş sayısını bulabiliriz.

```python
hotels.day_name.value_counts()
```

Eğer giriş yapılan güne karşılık gelen otelde kalış gününü görmek istersek "pazartesi girenler toplamda 5000 gün kalmış gibi" aşağıdaki kodu kullanabiliriz.

```python
hotels['total_stays'] = hotels['stays_in_weekend_nights'] + hotels['stays_in_week_nights']
```

```python
hotels2 = hotels.groupby('day_name')[['total_stays']].sum()
hotels2
```
